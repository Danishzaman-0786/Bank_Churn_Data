```python
import pandas as pd
v=pd.read_csv("/content/Churn_Modelling.csv")
```

```python
v.drop(columns=["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
v = pd.get_dummies(v, columns=["Geography", "Gender"],drop_first=True)
```

```python
v
```

```python
v
```

```python
x=v.drop(columns=["Exited"],axis=1)
y=v["Exited"]
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split as t
x_train,x_test,y_train,y_test=t(x,y,test_size=.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler as s
z_score=s()

x1_train=z_score.fit_transform(x_train)
x1_test=z_score.transform(x_test)
```

```python
x1_train.shape
```

```python
#now we will aply the keras library
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#he make a a sequential perceptron
model=Sequential()
```

```python
model.add(Dense(3,activation="sigmoid",input_dim=11))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="Adam",metrics=["accuracy"])
```

```python
history=model.fit(x1_train,y_train,epochs=10,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
model.predict(x1_test)
```

```python
y_pred=model.predict(x1_test)
import numpy as np

y_pred1=np.where(y_pred>.5,1,0)
```

```python
from sklearn.metrics import accuracy_score as score
score(y_test,y_pred1)
```
